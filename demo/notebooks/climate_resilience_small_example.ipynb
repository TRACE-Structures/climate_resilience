### Download data tmax

```python
example_json = 'https://raw.githubusercontent.com/TRACE-Structures/climate_resal/main/demo/data/small_example.json'
df = crl.run_all_models(example_json,                                   # Linked json (reduced version of json)
                    "maximum_2m_temperature_in_the_last_24_hours")        # Variable: "maximum_2m_temperature_in_the_last_24_hours"
                                                                          #           "minimum_2m_temperature_in_the_last_24_hours"
                                                                          #           "mean_precipitation_flux"
                                                                          #           "10m_wind_speed"
```

### Processing Data

```python
dfs_concat = crl.concat_models_along_time(df)
```

```python
dfs_yearly = crl.compute_yearly_ext(dfs_concat)
```

```python
dfs_window = crl.split_time_windows(dfs_yearly, window_length=4, shift=2)
```

```python
dfs_means, dfs_cov, dfs_charact = crl.compute_window_results(dfs_window, T_r=50, variable="tasmax", method='LSM')
```

```python
dfs_fc = crl.compute_fc_from_charact(dfs_charact, variable="tasmax")
```

### Export Data

```python
crl.export_csv(dfs_charact, "charact", "tasmax_charact_LSM.csv")
```
